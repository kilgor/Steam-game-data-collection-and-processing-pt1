```python
#calling all the csv files in the folder

df_overwiew = pd.read_csv('../data/raw/Dummy files/overview.csv')
df_copies_sold = pd.read_csv('../data/raw/Dummy files/copies sold.csv')
df_revenue = pd.read_csv('../data/raw/Dummy files/revenue.csv')
df_top_seller_rank = pd.read_csv('../data/raw/Dummy files/top seller rank.csv')
df_ccu = pd.read_csv('../data/raw/Dummy files/CCU.csv')
df_price = pd.read_csv('../data/raw/Dummy files/price.csv')
df_followers = pd.read_csv('../data/raw/Dummy files/followers.csv')
df_review = pd.read_csv('../data/raw/Dummy files/review.csv')
df_average_playtime = pd.read_csv('../data/raw/Dummy files/average playtime.csv')
df_outstanding_wishlist = pd.read_csv('../data/raw/Dummy files/outstanding wishlist.csv')
df_positive_reviews = pd.read_csv('../data/raw/Dummy files/positive reviews.csv')
```

```python
# Empty all dataframes except df_price
df_overwiew = df_overwiew.iloc[0:0]
df_copies_sold = df_copies_sold.iloc[0:0]
df_revenue = df_revenue.iloc[0:0]
df_top_seller_rank = df_top_seller_rank.iloc[0:0]
df_ccu = df_ccu.iloc[0:0]
df_followers = df_followers.iloc[0:0]
df_review = df_review.iloc[0:0]
df_average_playtime = df_average_playtime.iloc[0:0]
df_outstanding_wishlist = df_outstanding_wishlist.iloc[0:0]
df_positive_reviews = df_positive_reviews.iloc[0:0]


```

```python
# Step 1: Convert 'x' (milliseconds) to a proper date format
df_price['x'] = pd.to_datetime(df_price['x'], unit='ms')

# Step 2: Create a date range from 01.01.2024 to 09.15.2024
date_range = pd.date_range(start='01.01.2024', end='09.15.2024')

# Step 3: Create a new DataFrame with this date range
df_full_dates = pd.DataFrame(date_range, columns=['x'])

# Step 4: Merge the new date range with df_price to fill in missing dates
df_price_full = pd.merge(df_full_dates, df_price, on='x', how='left')

# Step 5: Forward fill the prices for missing dates
df_price_full['y'] = df_price_full['y'].fillna(method='ffill')

# Step 6: Transpose (rotate) the DataFrame
df_price_rotate = df_price_full.set_index('x').T

# Step 7: Rename the index to match df_copies_sold style (first row with dates)
df_price_rotate.columns = df_price_full['x'].dt.strftime('%a %b %d %Y')

# Step 8: Reset the index to remove the row and column headers (clean up x and y)
df_price_rotate.index = [0]

# Preview the rotated and cleaned dataframe
df_price_rotate.head()
```

```python
# Reset the index to ensure 'x' column is removed
df_price_rotate.columns.name = None  # Removes the 'x' column header name
df_price_rotate.reset_index(drop=True, inplace=True)  # Reset index without adding new one

# Preview the cleaned dataframe
df_price_rotate.head()
```

```python
#remove index
df_price_rotate.reset_index(drop=True, inplace=True)
```

```python
#fill with 0
df_price_0 = df_price_rotate.fillna(0)
```

```python
#save to csv
df_price_0.to_csv('../data/raw/Dummy files/price_0.csv', index=False)
```

```python
#befor I run this code I pip installed openpyxl as: pip install openpyxl

#I ran this code 2 times. first itentify the missing files 
#and run again to create 2nd excel file with name missing_files_report_2.xlsx to ensure if there are any missing files left

# Define the main library path and the required file names
main_library_path = '../data/raw/game lib'
required_files = [
    'overview.csv', 'copies sold.csv', 'revenue.csv', 'top seller rank.csv', 
    'CCU.csv', 'price.csv', 'review.csv', 'followers.csv', 
    'average playtime.csv', 'outstanding wishlist.csv', 'positive reviews.csv'
]

# Create an empty list to store results
missing_files_report = []

# Loop through each folder in the main library
for folder_name in os.listdir(main_library_path):
    folder_path = os.path.join(main_library_path, folder_name)
    
    # Check if it's a directory (to skip any non-folder files)
    if os.path.isdir(folder_path):
        # List all the files in the folder
        folder_files = os.listdir(folder_path)
        
        # Check for missing files
        missing_files = [file for file in required_files if file not in folder_files]
        
        # If there are missing files, record them
        if missing_files:
            missing_files_report.append({
                'Folder Name': folder_name,
                'Missing Files': ', '.join(missing_files)
            })

# Convert the report to a DataFrame
df_missing_files = pd.DataFrame(missing_files_report)

# Save the result to an Excel file
output_path = '../data/interim/missing_files_report_2.xlsx'
df_missing_files.to_excel(output_path, index=False)

print("Missing files report generated and saved to Excel.")
```
